```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'    # windows 한글폰트 연결
# matplotlib.rcParams['font.family'] = 'AppleGothic'      # mac 한글폰트 연결
matplotlib.rcParams['font.size'] = 15                   # 폰트 사이즈
matplotlib.rcParams['axes.unicode_minus'] = False       # 그래프 마이너스(-) 표시
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)                 # plot(x축 데이터, y축 데이터)
plt.title('온도조절 그래프')     # 상단 제목 입력
# x축 설명문구, color : 색상 지정(색상값/16진수코드/약식색상), loc = right, left, center(default)
plt.xlabel('X축 설명 - 월별', color='red', loc='right')
# x축 설명문구, color : 색상 지정, loc = top, bottom, center(default)
plt.ylabel('y축 설명 - 월별', color='#00aa00', loc='top')
# plt.ylabel('y축 설명 - 월별', color='indigo', loc='top')
# plt.ylabel('y축 설명 - 월별', color='g')     # y축 설명문구
# plt.xlabel('X축 설명 - 월별')   # x축 설명문구
# plt.ylabel('y축 설명 - 온도')   # y축 설명문구
plt.xticks([1, 2, 3])       # x축 표시 간격
plt.yticks([2, 4, 6, 8, 10, 12, 16])       # y축 표시 간격
# plt.yticks([2, 4, 6, 8])       # y축 표시 간격
# plt.xticks([1, 1.2, 1.5, 2, 3])       # x축 표시 간격
plt.show()
```
