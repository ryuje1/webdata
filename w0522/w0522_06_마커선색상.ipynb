```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'    # windows 한글폰트 연결
# matplotlib.rcParams['font.family'] = 'AppleGothic'      # mac 한글폰트 연결
matplotlib.rcParams['font.size'] = 15                   # 폰트 사이즈
matplotlib.rcParams['axes.unicode_minus'] = False       # 그래프 마이너스(-) 표시
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
df = pd.read_excel('data/score.xlsx', index_col='지원번호')
df
```

```python
# label : 범례
# 선형태 - linewidth : 선의 두께
# ls(linestyle) : None 선제거, : 점선, -- : 대쉬선, -. 점선점, color 선 색상
# marker : 지점에 마커 표시 - o, v, x, s
# mec(markeredgecolor): 마커 테두리 색상, mfc(markerfacecolor) : 마커 색상
## color, 마커모양, 라인스타일
plt.plot(x, y, 'rv:', label='서울', mec='blue', mfc='yellow')          # 'ro--' : color r, 마커 모양 o, 라인스타일 --
# plt.plot(x, y, 'ro--', label='서울')          # 'ro--' : color r, 마커 모양 o, 라인스타일 --
# plt.plot(x, y, label='서울', linewidth=2, ls='-.', color='red', alpha=0.7,
#          marker='o', markersize=10, mec='red', mfc='yellow')          # 꺾은선 그래프
# plt.plot(x, y, label='서울', linewidth=2, marker='x', linestyle='None')          # 꺾은선 그래프
# plt.plot(df['이름'], df['국어'])          # 리스트 입력 시 그래프 나타남
plt.title('그래프')     # 한글지원부분 설정해야 깨지지 않고 출력
# loc 범례 위치 - x축 : left, center, right / y축 : upper, center, lower 
# 위치점 지정 loc=(0.5, 0.5)    # 가운데지점
# plt.legend(loc=(0.25, 0.5))
plt.legend()
plt.show()             # 실행구문 제거
```

```python
plt.plot(df['이름'], df['국어'], label='국어', marker='o')
plt.plot(df['이름'], df['영어'], label='영어', marker='v', alpha=0.2)
plt.plot(df['이름'], df['수학'], label='수학', marker='X', alpha=0.2)
plt.title("학생성적그래프")
plt.legend()
plt.show()
```
