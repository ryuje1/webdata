```python
name = ['홍길동', '유관순', '이순신']
# no = [1, 2, 3]        # 리스트 타입
no = np.arange(1, 4)    # numpy 배열 형태로 적용

kor = [100, 80, 90]
eng = [95, 85, 92]
math = [80, 90, 77]
```

```python
colors = ['r', 'g', 'b']
### 바의 사이즈 조절 width
### no numpy 배열 형태
w = 0.2
# bar(x좌표, y좌표)
bar = plt.bar(no-w, kor, width=0.2)
bar[0].set_hatch('/')   # bar 무늬 적용
bar[1].set_hatch('x')   # bar 무늬 적용
bar[2].set_hatch('..')   # bar 무늬 적용
plt.bar(no, eng, width=0.2)
plt.bar(no+w, math, width=0.2)
### 각각의 막대그래프 위에 점수를 출력하시오.
for i, v in enumerate(kor):
    plt.text(no[i]-w, kor[i]+1, v, ha='center')
    plt.text(no[i], eng[i]+1, eng[i], ha='center')
    plt.text(no[i]+w, math[i]+1, math[i], ha='center')
plt.ylim(70, 105)
plt.xticks(no, name)        # x 좌표 이동에 필요한 수치 값을 no를 사용할 수 있음

plt.show()
```

```python
name = ['홍길동', '유관순', '이순신']
# no = [1, 2, 3]        # 리스트 타입
no = np.arange(1, 4)    # numpy 배열 형태로 적용

kor = [100, 80, 90]
eng = [95, 85, 92]
math = [80, 90, 77]
```

```python
# 3개의 막대 그래프를 출력하시오. grid : --, 색상 : 검정, 투명도 : 0.1
# 3개 막대 그래프 위에 점수를 출력하시오.

plt.bar(no-0.2, kor, width=0.2)
plt.bar(no, eng, width=0.2)
plt.bar(no+0.2, math, width=0.2)
plt.xticks(no, name)
plt.ylim(70, 105)
plt.grid(linestyle='--', color='black', alpha=0.1)
for i, v in enumerate(kor):
    plt.text(no[i]-0.2, kor[i]+1, v, ha='center')
    plt.text(no[i], eng[i]+1, eng[i], ha='center')
    plt.text(no[i]+0.2, math[i]+1, math[i], ha='center')
```
