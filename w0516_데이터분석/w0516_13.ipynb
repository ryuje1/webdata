```python
df.columns

# '종목명', '현재가', '등락률', '시가총액' 출력하시오.
# 10-20까지 출력하시오.
# df[['종목명', '현재가', '등락률', '시가총액']][10:20]

## columns 3-5 출력
# df[df.columns[3:6]][3:20]       # column 기준으로 row 출력

## rows 3:30 출력
# df[3:30]

### 종목명을 가지고 역순정렬 상위 10개까지 출력하시오.
df.index                                        # index 확인
df.reset_index(inplace=True)                    # index 초기화
df.set_index("종목명", inplace=True)             # index 종목명으로 변경
# df.sort_index(inplace=True)                     # 순차정렬
df.sort_index(ascending=False, inplace=True)    # 역순정렬
df.head(10)                                     # 상위 10개 출력
df[df.columns[[1,3,-1]]][:5]                    # 컬럼 1, 3, -1(끝번) 출력 / 상위 5개 출력
```

```python
#### df = column 정보를 출력 / df.iloc, df.loc = row 정보를 출력
df['현재가']            # column에 대한 출력 (row 정보가 출력되지만 row에 대한 출력X)
df[:5]                 # df[:][:5] 에서 [:] 생략
df.iloc[:5]            # row에 대한 출력
```

```python
df.info()           # 현재가 타입 - object
df['현재가'][:3]     # 타입 변경 필요

## 숫자타입 변경시 천단위 표시 제거
# replace : str.replace(",", "") => replace는 inplace=True 불가
# 타입 변경 : astype(int)
df['현재가'] = df['현재가'].str.replace(",", "").astype(int)
df['현재가']    # object

df.info()

### 최대값, 최소값, 상위 5개 출력, 하위 5개 출력, 평균
df.describe()
df['현재가'].max()
df['현재가'].min()
df['현재가'].mean()
df['현재가'].head(5)
df['현재가'].tail(5)
```
