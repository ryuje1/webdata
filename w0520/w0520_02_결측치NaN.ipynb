```python
### 국어 점수를 모두 NaN 처리 후 50으로 모두 변경하시오.
# df['국어'] = np.nan
# df['국어'].fillna(50)

### 1. 국어 평균점수를 먼저 구하고, 2. 국어 점수 모두 NaN처리, 3. 평균점수로 변경
# 국어 점수를 NaN처리 후 이전 국어점수의 평균점수로 변경하시오.
df = pd.read_excel('data/score.xlsx')
kor = df['국어'].mean()       # 62.5
df['국어'] = np.nan
df['국어'].fillna(kor, inplace=True)
df
```
