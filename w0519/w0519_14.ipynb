```python
import pandas as pd

data = {
   '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
   '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
   '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
   '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)
```

```python
# 평점이 9.0 이상이거나 관객 수가 1700명 이상인 영화를 출력하시오.
df.set_index('영화', inplace=True)
df[(df['평점']>=9.0) | (df['관객 수']>=1700)]
```

```python
### startswith 함수 - 해당 글자로 시작되는 문자 검색
df = pd.read_csv('data/stuscore.csv', index_col='번호')
filt = df['이름'].str.startswith('S')
df[filt]
```

```python
### endswith 함수 - 해당 글자로 끝나는 문자 검색
df = pd.read_csv('data/stuscore.csv', index_col='번호')
df[df['이름'].str.endswith('덕')]
```

```python
### contains 함수 - 해당 글자가 포함되어있는 문자 검색
df[df['이름'].str.contains('s')]
df[~df['이름'].str.contains('s')]       # ~ : not을 의미
```

```python
### a가 들어가 있는 문자를 검색하시오.
df[df['이름'].str.contains('a')].count()        # 40
df[df['이름'].str.contains('A')].count()        # 8
df[(df['이름'].str.contains('a')) | (df['이름'].str.contains('A'))].count()       # 48
```

```python
df['이름'].str.lower()      # 모든 글자를 소문자로 변경
# 모든 글자를 소문자로 변경 후 a가 포함되어있는 문자 검색
df[df['이름'].str.lower().str.contains('a')].count()        # 48

df['이름'].str.upper()      # 모든 글자를 대문자로 변경
# 모든 글자를 대문자로 변경 후 A가 포함되어있는 문자 검색
df[df['이름'].str.upper().str.contains('A')].count()        # 48
```
