데이터 선택 loc - row 행에 대한 선택

```python
import pandas as pd
df = pd.read_excel('data/score.xlsx', index_col='지원번호')
df
```

```python
### 컬럼부분
df['이름']
df[['이름', '학교']]
df.columns
df[df.columns[:3]][1:3]
```

```python
df.index
```

loc 행 출력 : index 이름을 가지고 출력

```python
### row(행) 부분 출력
# loc[] : 쉼표(,)로 앞쪽 행, 뒤쪽 열 의미

df.loc['1번']
df.loc[['1번', '2번']]      # df[0:2] 와 동일
df[0:2]

### row, column 부분 출력
df.loc['1번']['이름']                  # 1번의 이름
df.loc['1번', '이름']
df.loc[['1번', '2번', '3번']]['이름']   # 1,2,3번의 이름
df.loc[['1번','2번'],['국어', '영어']]  # 1,2번의 국어, 영어 점수
df.loc['1번', ['국어', '영어']]
# df.iloc[[df.index[0:1]], '학교']

### row, column 슬라이싱 출력
df.loc['1번':'5번', '국어':'사회']      # 1,2,3,4,5번의 국어부터 사회까지 점수
```

iloc 행 출력 : 순번을 가지고 출력


```python
# 행 출력
df.iloc[0]
df.iloc[4]
df.iloc[0:4]

# 행의 기준으로 열을 출력 - 1행(1번 학생) 가져와서 2번째(학교) 컬럼 데이터 출력
df.iloc[0, 1]           # df.loc['1번', '학교'] 와 동일
df.iloc[[0, 1], 1]   # df.loc[['1번', '2번'], '학교'], df.loc[['1번', '2번'], df.columns[1]] 와 동일
df.iloc[[0, 1, 3], 1]   # df.loc[['1번', '2번', '4번'], '학교'] 와 동일

df.iloc[[0,1], [3,4]]   # 0,1 행, 3, 4열 출력


# 슬라이싱 출력
df.iloc[:4, 3:]
df.iloc[1:4, 3:7]       # 1, 4행 앞까지 3행 출력, 3열 8열 앞까지 4개 출력
```

```python
### 3, 4, 5행 출력
df.loc[['3번','4번','5번']]
df.loc['3번':'5번']
df.iloc[2:5]
df.iloc[[2,3,4]]

### 1, 5, 6, 8행 출력
df.loc[['1번', '5번', '6번', '8번']]
df.iloc[[0, 4, 5, 7]]

## 4, 5행 학교, 키 출력
df.loc[['4번', '5번']][['학교', '키']]
df.loc[['4번', '5번'], ['학교', '키']]
df.loc['4번':'5번', '학교':'키']
df.iloc[[3, 4]][['학교', '키']]
df.iloc[[3, 4], [1, 2]]
df.iloc[3:5, 1:3]
```
