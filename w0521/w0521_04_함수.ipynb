```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('data/score.xlsx', index_col='지원번호')
```

```python
# int 타입에 cm 문자 추가방법
# df['키'] = df['키'].astype(str) + 'cm'

# 함수 사용해서 추가
# def add_cm(height):
#     return str(height)+'cm'
# df['키'] = df['키'].apply(add_cm)   # 키 각각의 데이터 값을 함수에 대입

df['합계'] = df['국어'] + df['영어'] + df['수학'] + df['사회'] + df['과학']
df['평균'] = df['합계']/5
df['등급'] = 'F'

# 함수를 사용해서 90점 이상 합격/90점 미만 불합격
# df['등급'] = '불합격'
# df.loc[df['평균'] > 90.0, '등급'] = '합격'

def rank(score):
    result = ""
    if score >=90: result = "합격"
    else: result = "불합격"
    return result
df['등급'] = df['평균'].apply(rank)
df
```

```python
### 90점 이상 A, 80점 이상 B, 70점 이상 C, 60점 이상 D, 그 이하 F
def rank(score):
    result=""
    if score >=90: result = 'A'
    elif score >=80: result = 'B'
    elif score >=70: result = 'C'
    elif score >=60: result = 'D'
    else: result = 'F'
    return result
df['등급'] = df['평균'].apply(rank)
df
```

```python
### 함수를 사용하여 SW특기 컬럼의 글자를 전부 대문자로 변경하고 NaN데이터는 C를 입력하시오.
# if pd.null()     # null 검색
df['SW특기'].str.upper()
df.fillna('C')

## isna(), isnull(), notnull()
def changeData(a_data):
    # null이 아니면
    # if pd.notnull(a_data): return a_data.upper()
    # else: return 'C'
    
    # null이면
    if pd.isnull(a_data): 'C'
    else: return a_data.upper()
df['SW특기'] = df['SW특기'].apply(changeData)
# df['등급'] = df['등급'].apply(changeData)     # 등급 소문자 변경
df
```

```python
### 함수를 사용해서 요일이  0:일요일 / 1:월 / 2:화 / 3:수 / 4:목 / 5:금 / 6:토 변경하시오.
def week(day):
    result = ""
    if day == 0: result = '일요일'
    elif day == 1: result = '월요일'
    elif day == 2: result = '화요일'
    elif day == 3: result = '수요일'
    elif day == 4: result = '목요일'
    elif day == 5: result = '금요일'
    else: result = '토요일'
    return result
df['요일'] = df['요일'].apply(week)
df
# df.info()
```
