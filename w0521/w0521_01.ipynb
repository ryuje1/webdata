```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('data/score.xlsx', index_col='지원번호')
```

```python
### numpy 명령어 np.nan : NaN데이터로 변경
df['학교'] = np.nan
```

```python
# 모든 NaN데이터에 적용
df.fillna("없음")       # inplace=True
df['학교'].fillna('가산고')
df['SW특기'].fillna('C')
```

```python
### dropna : NaN데이터 삭제 - row기준으로 삭제
# df.dropna()
# axis 기준 : index, columns / how : all (모두 NaN이면 삭제), any(NaN이 1개라도 있으면 삭제, default)
# df.dropna(axis='index', how='all')
# df.dropna(axis='columns', how='all')
```

```python
### 데이터 정렬 sort_index, sort_values
# index 기준으로 정렬 - ascending : True 순차정렬, False 역순정렬
df.sort_index(ascending=False)

# index 삭제
df.reset_index()        # drop=True : 인덱스 삭제, inplace=True : 반영

# index 지정
# df.set_index('이름', inplace=True)

# index 정렬
df.sort_index(ascending=False)
```

```python
# 각 컬럼별 정렬
# sort_values
df.sort_values('키')    # inplace=True
df.sort_values('키', ascending=False)   # inplace=True
df.sort_values(['국어', '영어'], ascending=[True, False])         # 국어 기준 1차 정렬, 국어 점수가 동일 점수일때 영어 기준 2차 정렬
```
