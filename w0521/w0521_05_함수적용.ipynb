```python
import pandas as pd
import numpy as np
```

```python
data = {
    '번호':['1번','2번','3번','4번','5번','6번','7번','8번'],
    '이름':['홍길동','유관순','이순신','강감찬','김구','김유신','홍길자','홍길순'],
    '달러':['$19,700','$18,400','$16,800','$18,700','$18,800','$20,200','$18,800','$19,000'],
    '우대':[90,40,80,40,15,80,55,100]
}
```

```python
df = pd.DataFrame(data)
df.set_index('번호', inplace=True)
df
```

```python
## 금액 : 달러 + 달러*우대 - 금액 컬럼을 추가하시오.
df['달러'] = df['달러'].str.replace("$", "").str.replace(",", "").astype(int)
# df['달러'] = df['달러'].str.replace(",", "").str.slice(1).astype(int)
df['금액'] = df['달러'] + df['달러']*df['우대']
df
```

```python
### 우대 50인 사람만 적용해서 금액을 산출
### 우대가 50 미만 달러 입력, 50 이상은 달러+달러*우대*0.01
df['금액'] = 0
df.loc[df['우대']>50, '금액'] = df['달러']+df['달러']*df['우대']*0.01
df.loc[~(df['우대']>50), '금액'] = df['달러']
df
```

```python
### 함수적용 50 이상 100, 50 미만 50
# 1. 컬럼 1개
# def change(data):
#     if data>=50: return 100
#     else: return 50
    
# df['금액'] = df['우대'].apply(change)
# df

# 2. 컬럼 2개
def change(data):
    if data['우대']>=50: return data['달러']+data['달러']*data['우대']*0.01
    else: return data['달러']
df['금액'] = df.apply(change, axis=1)   # axis=1 : x축. 행(가로)의 데이터를 보냄
df
```
