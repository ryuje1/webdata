```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('data/score.xlsx', index_col='지원번호')
```

```python
## 그룹화
df.groupby('학교').get_group('가산고')
df.groupby('학교').get_group('디지털고')
df.groupby('학교').size()                   # size 개수
df.groupby('학교').size()['가산고']          # 가산고 학생수 - 5
df.groupby('학교').size()['디지털고']        # 디지털고 학생수 - 3
df.groupby('학교')['키'].mean()
df.groupby('학교')[['키', '국어', '영어']].mean()
df.groupby('학교')[['키', '국어', '영어']].max()
df.groupby('학교')[['키', '국어', '영어']].min()
df.groupby('학교')[['키', '국어', '영어']].sum()
# df.groupby('학교')['이름'].mean()            # 숫자여야 가능 => 에러
```

```python
# df.groupby('학교').mean()           # mean() => 숫자가 아닌 타입이 있을 경우 에러
df.groupby('학교').count()          # count() => 숫자가 아닌 타입이 있어도 가능
df['학교'].count()                  # 학교에 대한 개수
df.groupby('학교')['학교'].count()   # 숫자가 아닌 타입이 있어도 가능
df['학교'].value_counts()           # 학교 그룹별 개수가 나타남. groupby().count()와 동일

# 그룹핑 개수
df.groupby('학교')['학년'].count()
df.groupby('학교')['학년'].count().loc['가산고']    # 가산고만 출력

# 1차 학교로 그룹화 후 2차 학년으로 그룹화 실행
df.groupby('학교')['학년'].value_counts()          # 학교별 학년의 개수
df.groupby(['학교', '학년'])['학년'].count()        # 위와 동일
```

```python
# df.groupby('학교')[['이름', 'SW특기']].count()
# df.groupby('학교').value_counts()
df['학교'].value_counts()
df.groupby('학교')['학년'].value_counts()
# agg() : 기본적인 통계정보를 알려줌.  - 평균, 최소값, 최대값, 합계
df.groupby('학교')['키'].agg(['mean', 'min', 'max', 'sum'])     # 학교 그룹화한 키의 평균
df['키'].agg(['mean'])                                          # 전체 키의 평균
df.groupby('학년')['키'].agg(['mean'])                          # 학년별 키의 평균
```

```python
df['성별'] = ['남', '여', '남', '남', '남', '남', '여', '여']
df
```

```python
#### 합계 컬럼 추가해서
# 학교에 성별로 합계컬럼의 평균, 최소, 최대, 합계 점수를 구하시오.
df['합계'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
df.groupby('학교')['성별'].value_counts()
df.groupby(['학교', '성별'])['합계'].agg(['mean', 'min', 'max', 'sum'])
df.groupby(['학교', '성별'])['합계'].mean()
df.groupby(['학교', '성별'])['합계'].min()
df.groupby(['학교', '성별'])['합계'].max()
df.groupby(['학교', '성별'])['합계'].sum()
```
