```python
import pandas as pd
import numpy as np
```

```python
data = {
   '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
   '개봉 연도' : [2014, 2019, 2017, 2014, 2019, 2012, 2014, 2017],
   '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
   '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)
```

```python
## 개봉연도 -> 2014년도 글자 추가해서 출력
df['개봉연도'] = df['개봉 연도'].astype(str) + '년도'
df

## 개봉연도별 관객수 평균, 최소, 최대, 합계를 구하시오.
df.groupby('개봉연도')['관객 수'].mean()
df.groupby('개봉연도')['관객 수'].min()
df.groupby('개봉연도')['관객 수'].max()
df.groupby('개봉연도')['관객 수'].sum()
df.groupby('개봉연도')['관객 수'].agg(['mean', 'min', 'max', 'sum'])
```
